```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
train_df =pd.read_csv("data/train.csv", parse_dates=['timestamp'])
```

```python
train_df.info()
```

```python
missing =train_df.isnull().sum()

missing.sort_values(ascending=False, inplace=True)

missing_real =missing[missing.values >0]

plt.figure(figsize=(8, 5))

missing_real.plot.bar(colormap ='Dark2')
```

```python
missing_real.index
```

```python
missing_real_percent =missing_real / train_df.shape[0]

plt.figure(figsize=(8, 5))

print(missing_real_percent.plot.bar(colormap ='gray'))
```
